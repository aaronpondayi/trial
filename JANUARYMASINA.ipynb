```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy 
import matplotlib.pyplot as plt
%matplotlib inline
#df = pd.read_csv("masina1.csv")
df =pd.read_csv("masina111.csv")
df.head()
df.info()
```

```python
plt.scatter(df['date'],df['readings'], color = 'green')
plt.title("Aaron Pondayi Masina Glucose Readings")
#plt.xlabel("Readings")
plt.xlabel("Days")
plt.ylabel("Readings")
#plt.ylabel("Days")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Days")
plt.ylabel("Readings")
#create basic scatterplot
plt.plot(df['date'],df['readings'] ,'o', color='red')

#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(df['date'],df['readings'], 1)

#add linear regression line to scatterplot 
plt.plot(df['date'], m*df['date']+b)
```

```python
c = np.corrcoef(df['date'],df['readings'])
print(' Correlation on data \n', c)
```
